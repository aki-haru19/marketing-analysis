```python
conversion_counts = df['Conversion'].value_counts()
conversion_rate = df['Conversion'].value_counts(normalize=True) * 100
print("Conversion counts:",conversion_counts, "Conversion rate:",conversion_rate)
```

```python
cr_cnt = conversion_counts.plot(kind='bar', color = 'salmon')
plt.title('Conversion counts')
plt.ylabel('counts')
plt.xticks(rotation = 0)
for container in cr_cnt.containers:
    cr_cnt.bar_label(container, fmt='{}') 
```

```python
cac_by_chanel = df.groupby('CampaignChannel')['Conversion'].sum()
print(cac_by_chanel)
cac_by_type = df.groupby('CampaignType')['Conversion'].sum()
print(cac_by_type)
```

```python
cac = df['AdSpend'].sum()/df['Conversion'].sum()
print(cac)
```

```python
#cac by CampaignChannel
channels = df.groupby('CampaignChannel').agg({
    'AdSpend': 'sum',
    'Conversion': 'sum'
}).reset_index()
channels['cac'] = channels['AdSpend']/channels['Conversion']
print(channels)
plot_channels = channels['cac'].plot(kind='bar', color = 'coral')
plt.title('CAC by CampaignChannel')
plt.ylabel('cac')
plt.xticks(rotation = 0)
for container in plot_channels.containers:
    plot_channels.bar_label(container, fmt='%.1f') 
```

```python
#cac by CampaignType
types = df.groupby('CampaignType').agg({
    'AdSpend': 'sum',
    'Conversion': 'sum'
}).reset_index()
types['cac'] = types['AdSpend']/types['Conversion']
print(types)
plot_types = plt.bar(types['CampaignType'], types['cac'], 
               color='#9b59b6', alpha=0.8)
plt.title('CAC by CampaignType ')
plt.ylabel('cac')
plt.xticks(rotation = 0)
```

```python
# Conversion rate by channel
c_r = df.groupby('CampaignChannel').agg({
    'Conversion': ['sum', 'count']
}).reset_index()
c_r['rate'] = (c_r[('Conversion','sum')] / c_r[('Conversion','count')])*100
c_r = c_r.sort_values('rate', ascending=False)
print(c_r)
plot_cr = plt.bar(c_r['CampaignChannel'], c_r['rate'], 
               color='sienna', alpha=0.8)
plt.title('Conversion Rate by channel ')
plt.ylabel('rate (%)')
plt.xticks(rotation = 0)
for i, value in enumerate(c_r['rate']):
    plt.text(i, value + 0.5, f'{value:.1f}%', ha='center')
```

```python
aov = 7000  #AOV (Average Order Value):** $7,000 (estimated, would come from transaction data in production)
```

```python
# roi 

df['revenue'] = aov * df['Conversion']
df['roi']= ((df['revenue'] - df['AdSpend']) / df['AdSpend'])*100
df['roi'].describe()
```

```python
roi_channel = df.groupby('CampaignChannel').agg({
    'Conversion': 'sum', 
    'AdSpend': 'sum' }).reset_index()
roi_channel['revenue'] = aov * roi_channel ['Conversion']
roi_channel['roi']= ((roi_channel ['revenue'] - roi_channel ['AdSpend']) / roi_channel ['AdSpend'])*100
print(roi_channel)
roi_bar = plt.bar(roi_channel['CampaignChannel'], roi_channel['roi'], alpha=0.8)

plt.title('ROI by Campaign Channel', fontsize=14, fontweight='bold')
plt.xlabel('Campaign Channel')
plt.ylabel('ROI (%)')
plt.title('ROI by Campaign Channel', fontsize=14, fontweight='bold')
plt.xlabel('Campaign Channel')
plt.ylabel('ROI (%)')
for bar, val in zip(roi_bar, roi_channel['roi']):
    plt.text(bar.get_x() + bar.get_width()/2, val + 0.3, 
             f'{val:.0f}%', ha='center', fontweight='bold', fontsize=9)
```

```python
# cac by channel/type
cac_all= df.groupby(['CampaignChannel', 'CampaignType']).agg({
    'Conversion': 'sum', 
    'AdSpend': 'sum' }).reset_index()
cac_all['cac'] = cac_all['AdSpend']/cac_all['Conversion']
pivot_data = cac_all.pivot(index='CampaignChannel', columns='CampaignType', values='cac')
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_data, annot=True, fmt='.2f', cmap='RdYlGn_r', 
            cbar_kws={'label': 'CAC ($)'}, linewidths=0.5)
plt.title('CAC: Campaign Channel vs Campaign Type', fontsize=14, fontweight='bold')
plt.xlabel('Campaign Type')
plt.ylabel('Campaign Channel')
plt.tight_layout()
plt.show()
```

```python
df['Email_Engagement'] = 'None'
df.loc[df['EmailOpens'] > 0, 'Email_Engagement'] = 'Opened'
df.loc[df['EmailClicks'] > 0, 'Email_Engagement'] = 'Clicked'
print(df['Email_Engagement'].value_counts())
# CR  engagement
email_cr = df.groupby('Email_Engagement')['Conversion'].mean() * 100
print(email_cr)
plot_email = plt.bar(email_cr.index, email_cr.values, color='green', alpha=0.8)
plt.title('Conversion Rate by email ')
plt.ylabel('rate (%)')
plt.xticks(rotation = 0)
for i, value in enumerate(email_cr.values):
    plt.text(i, value + 0.5, f'{value:.1f}%', ha='center')
```

```python
#ctr 
ctr = df.groupby('Conversion')['ClickThroughRate'].mean()

print("Average CTR:",df['ClickThroughRate'].mean())
print(f"Not converted (0): {ctr[0]:.4f}")
print(f"Converted (1): {ctr[1]:.4f}")

corr_ctr = df['ClickThroughRate'].corr(df['Conversion'])
print(corr_ctr)

#scatter here looks awful 
```

```python

df['agerange'] = pd.cut(df['Age'], 
                          bins=[0, 25, 35, 45, 55, 65, 100], 
                          labels=['18-25', '26-35', '36-45', '46-55', '56-65', '65+'])
print(df['agerange'].value_counts())
age_cr = df.groupby('agerange')['Conversion'].mean() *100
age_cr = age_cr.sort_values(ascending = False)
print(age_cr)
plot_age = plt.bar(age_cr.index, age_cr.values, color='violet', alpha=0.8)
plt.title('Conversion Rate by age ')
plt.ylabel('rate (%)')
plt.xticks(rotation = 0)
for i, value in enumerate(age_cr.values):
    plt.text(i, value + 0.5, f'{value:.1f}%', ha='center')
```

```python
print("======================== KEY INSIGHTS ===========================")

# 1. BEST CHANNEL
best_channel_cac = channels.loc[channels['cac'].idxmin()]
best_channel_roi = roi_channel.loc[roi_channel['roi'].idxmax()]

print("\n1) BEST PERFORMING CHANNEL:")
print(f"   Channel: {best_channel_cac['CampaignChannel']}")
print(f"   CAC: ${best_channel_cac['cac']:.2f} (lowest)")
print(f"   ROI: {best_channel_roi['roi']:.1f}% (highest)")
print(f"   Conversions: {best_channel_cac['Conversion']:.0f}")
print(f"    Why: Most cost-effective acquisition + highest return")

# 2. WORST CHANNEL
worst_channel_cac = channels.loc[channels['cac'].idxmax()]
worst_channel_roi = roi_channel.loc[roi_channel['roi'].idxmin()]

print("\n2) UNDERPERFORMING CHANNEL:")
print(f"   Channel: {worst_channel_cac['CampaignChannel']}")
print(f"   CAC: ${worst_channel_cac['cac']:.2f} (highest)")
print(f"   ROI: {worst_channel_roi['roi']:.1f}% (lowest)")
print(f"    Gap: {((worst_channel_cac['cac'] - best_channel_cac['cac'])/best_channel_cac['cac']*100):.1f}% more expensive than best channel")
print(f"    Recommendation: Optimize or reduce budget")

# 3. CHANNEL RANKING
print("\n3) CHANNEL PERFORMANCE RANKING (by ROI):")
channel_summary = roi_channel.merge(channels[['CampaignChannel', 'cac']], on='CampaignChannel')
channel_summary = channel_summary.sort_values('roi', ascending=False)
channel_summary = channel_summary.reset_index(drop=True)
for idx, row in channel_summary.iterrows():
    print(f"   {idx+1}. {row['CampaignChannel']:13s} | CAC: ${row['cac']:7.2f} | ROI: {row['roi']:5.1f}%")

# 4. CAMPAIGN TYPE
best_type = types.loc[types['cac'].idxmin()]
worst_type = types.loc[types['cac'].idxmax()]

print("\n4) CAMPAIGN TYPE PERFORMANCE:")
print(f"   Best: '{best_type['CampaignType']}' (CAC: ${best_type['cac']:.2f})")
print(f"   Worst: '{worst_type['CampaignType']}' (CAC: ${worst_type['cac']:.2f})")
print(f"    Gap: {((worst_type['cac'] - best_type['cac'])/best_type['cac']*100):.1f}% difference")

# 5. ROI ANALYSIS
avg_roi = roi_channel['roi'].mean()
print("\n5) ROI ANALYSIS:")
print(f"   Average ROI: {avg_roi:.1f}%")
print(f"   Best: {roi_channel['roi'].max():.1f}% ({roi_channel.loc[roi_channel['roi'].idxmax(), 'CampaignChannel']})")
print(f"   Worst: {roi_channel['roi'].min():.1f}% ({roi_channel.loc[roi_channel['roi'].idxmin(), 'CampaignChannel']})")
print(f"    Spread: {roi_channel['roi'].max() - roi_channel['roi'].min():.1f}% between best and worst")

# 6. EMAIL MARKETING
print("\n6) EMAIL MARKETING INSIGHT:")
print(f"   Clicked: {email_cr['Clicked']:.1f}% CR")
print(f"   Opened only: {email_cr['Opened']:.1f}% CR")
print(f"   No engagement: {email_cr['None']:.1f}% CR")
print(f"    Gap: {email_cr['Clicked'] - email_cr['Opened']:.1f}% between clicks and opens")

# 7. AGE SEGMENTATION
best_age = age_cr.idxmax()
worst_age = age_cr.idxmin()
avg_cr = age_cr.mean()

print("\n7) AGE SEGMENTATION:")
print(f"   Best: {best_age} ({age_cr.max():.1f}% CR)")
print(f"   Worst: {worst_age} ({age_cr.min():.1f}% CR)")
print(f"   Average: {avg_cr:.1f}% CR")
print(f"    Best group is {age_cr.max() - avg_cr:.1f}% above average")

# 8. CTR IMPACT
print("\n8) CLICK-THROUGH RATE IMPACT:")
print(f"   Converted: {ctr[1]:.4f} avg CTR")
print(f"   Non-converted: {ctr[0]:.4f} avg CTR")
print(f"   Difference: {((ctr[1] - ctr[0])/ctr[0]*100):.1f}% higher for converters")
print(f"   Correlation: {corr_ctr:.3f} (weak positive)")

# 9. OVERALL PERFORMANCE
print("\n9) OVERALL CAMPAIGN PERFORMANCE:")
total_spend = df['AdSpend'].sum()
total_conv = df['Conversion'].sum()
total_revenue = total_conv * aov
profit = total_revenue - total_spend

print(f"   Total Ad Spend: ${total_spend:,.0f}")
print(f"   Total Conversions: {total_conv:,.0f}")
print(f"   Overall CAC: ${total_spend/total_conv:.2f}")
print(f"   Conversion Rate: {df['Conversion'].mean()*100:.1f}%")
print(f"   Estimated Revenue: ${total_revenue:,.0f}")
print(f"   Estimated Profit: ${profit:,.0f}")
print(f"   Overall ROI: {(profit/total_spend)*100:.1f}%")
```
